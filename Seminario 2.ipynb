```python
import scipy.io.wavfile as wf
import numpy as np

# Lectura de archivo de audio
fs, s = wf.read("Datos/altura.wav")
# Separaci√≥n de audio en dos partes
mitad = int(len(s) / 2)
s1 = s[:mitad]
s2 = s[mitad:]
# Escritura de archivos de audio
wf.write("audioInicial.wav", fs, s1)
wf.write("audioFinal.wav", fs, s2)
```

```python
import numpy as np
import scipy.io.wavfile as wf
import matplotlib.pyplot as plt

fs, s = wf.read('Datos/ciudad.wav')
t = np.arange(len(s)) / fs

fig, ax = plt.subplots()
ax.plot(t, s)
ax.set(xlabel='Tiempo (s)')
```

```python
import scipy.io.wavfile as wf
import sounddevice as sd

# Leer el archivo de audio
fs_original, data = wf.read('Datos/mar.wav')

# Reproducir con la frecuencia de muestreo original
print("Reproduciendo con la frecuencia original...")
sd.play(data, fs_original)
sd.wait()

# Definir una nueva frecuencia de muestreo (por ejemplo, el doble)
fs_nueva = int(fs_original * 2)

# Reproducir con la nueva frecuencia de muestreo
print(f"Reproduciendo con una frecuencia de muestreo de {fs_nueva} Hz...")
sd.play(data, fs_nueva)
sd.wait()
```
