```python
import matplotlib.pyplot as plt

# Primera serie de datos
x1 = [
    8.5, 10.8, 12.4, 14.6, 16.2, 18.1, 19.3, 20.6, 21.8, 23.8,
    25.9, 27.8, 29.6, 31.4, 34.8, 37.3, 40.3,
]
y1_V = [
    3.9, 4.0, 3.92, 3.87, 4.4, 6.9, 9.96, 6.8, 4.12, 3.92,
    3.52, 2.72, 1.68, 0.71, 0.64, 0.48, 0.32,
]
y1 = [v / 100 for v in y1_V]  # Conversión a voltios


# Segunda serie de datos
x2 = [8, 10, 12.3, 14.6, 17.3, 19.7, 22.4, 25.5, 27.5, 30, 32.9, 34.1, 36.9, 39.1, 41.7]
y2_mV = [76, 82, 83, 70, 72, 80, 84, 92, 112, 116, 121, 121.36, 124, 144, 136]
y2 = [v / 1000 for v in y2_mV]  # Conversión a voltios


# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x1, y1, marker="o", linestyle="-", color="blue", label="Serie 1")
plt.plot(x2, y2, marker="s", linestyle="--", color="red", label="Serie 2")

# Detalles del gráfico
plt.title("Comparación de dos series de datos")
plt.xlabel("Temperatura [ºC]")
plt.ylabel("Voltaje [mV]")
plt.ylim(0, 0,15)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
