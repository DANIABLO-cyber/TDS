# Curvas P–I y ajustes lineales
___


```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------
# Sección 1: Curvas P–I y ajustes lineales
# ------------------------------------

# Eje X para 10 °C
a = np.arange(30, 66, 5)
b = np.arange(67, 70, 2)
c = np.arange(70, 72, 1)
d = np.arange(73, 80, 2)
e = np.arange(80, 83, 2)
f = np.arange(83, 86, 2)
x_10 = np.concatenate([a, b, c, d, e, f])

# Eje X para 30 °C (continúa desde x_10)
g = np.arange(87, 90, 2)
h = np.arange(90, 101, 2)
x_30 = np.concatenate([x_10, g, h])

# Eje Y (Voltaje en mV)
V_10 = np.array([
    4.9,   6.14,  7.08,  7.7565, 7.88,  9.32, 10.96, 13.56, 18.6,  22.8,
   27.9,  42.6, 103.0, 174.0, 208.0, 284.0, 311.0, 386.0, 410.0, 486.0
])
V_30 = np.array([
   4.633,  5.167,  5.427,  6.37,   6.88,   7.58,   8.573,  11.327, 12.1,   12.97,
  13.473, 14.4,   15.1,   17.6,   23.27,  38.4,   45.2,   79.07,  95.13,  111.67,
 134.933,151.33,174.67,188.67,230.67,271.33,306.0,342.67
])

# Figura 1: P–I a 10 °C y 30 °C
plt.figure()
plt.plot(x_10, V_10, 'o-', label='Temperatura 10 °C')
plt.plot(x_30, V_30, 'o-', label='Temperatura 30 °C')
plt.xlabel('Corriente (mA)')
plt.ylabel('Voltaje (mV)')
plt.title('Curvas P–I del diodo láser')
plt.grid(True)
plt.legend()

# Ajuste lineal para 30 °C (a partir del punto 15; índices Python empiezan en 0)
a30, b30 = 13.674, -1046.6
idx30 = np.arange(14, len(x_30))
y30 = a30 * x_30[idx30] + b30

# Figura 2: Regresión 30 °C
plt.figure()
plt.plot(x_30, V_30, 'o-', label='Datos 30 °C')
plt.plot(x_30[idx30], y30, '--', label='Ajuste lineal')
plt.xlabel('Corriente (mA)')
plt.ylabel('Voltaje (mV)')
plt.title('Regresión P–I a 30 °C')
plt.grid(True)
plt.legend()

# Ajuste lineal para 10 °C (a partir del punto 11)
a10, b10 = 30.569, -2125.8
idx10 = np.arange(10, len(x_10))
y10 = a10 * x_10[idx10] + b10

# Figura 3: Regresión 10 °C
plt.figure()
plt.plot(x_10, V_10, 'o-', label='Datos 10 °C')
plt.plot(x_10[idx10], y10, '--', label='Ajuste lineal')
plt.xlabel('Corriente (mA)')
plt.ylabel('Voltaje (mV)')
plt.title('Regresión P–I a 10 °C')
plt.grid(True)
plt.legend()
```

# REGRESION LINEAL
___

```python
import numpy as np
import matplotlib.pyplot as plt

# — Datos “con espejo” (21 puntos)
Temp_con = np.array([
  9.349, 10.715, 11.703, 13.345, 14.124, 16.946, 19.038, 21.792,
 24.064, 24.183, 24.892, 27.429, 29.468, 30.626, 30.871, 32.087,
 33.320, 34.833, 37.378, 38.870, 42.779
])
Amp_con = np.array([
 0.630, 0.496, 0.528, 0.462, 0.614, 0.894, 0.444, 0.102,
 0.091, 0.176, 0.027, 0.396, 0.807, 0.852, 1.000, 1.000,
 1.000, 1.000, 1.000, 0.986, 0.894
])

# — Datos “sin espejo” (17 puntos)
Temp_sin = np.array([
 10.098, 10.960, 13.567, 15.171, 16.410, 17.129, 20.315, 20.664,
 23.270, 24.739, 28.088, 31.302, 33.709, 35.461, 37.853, 38.153, 42.120
])
Amp_sin = np.array([
 0.797, 0.774, 0.966, 0.961, 1.000, 0.964, 0.970, 0.987,
 0.932, 0.660, 0.634, 0.460, 0.548, 0.297, 0.148, 0.010, 0.075
])

# 1) Selección de la región lineal (bombeo eficiente)
mask_con = (Temp_con >= 24) & (Temp_con <= 33)
mask_sin = (Temp_sin >= 10) & (Temp_sin <= 18)

x1, y1 = Temp_con[mask_con], Amp_con[mask_con]
x2, y2 = Temp_sin[mask_sin], Amp_sin[mask_sin]

# 2) Ajuste lineal mediante polyfit
a1, b1 = np.polyfit(x1, y1, 1)       # pendiente e intercepto "con espejo"
a2, b2 = np.polyfit(x2, y2, 1)       # pendiente e intercepto "sin espejo"

# 3) Cálculo de R² para evaluar bondad de ajuste
def r_squared(x, y, a, b):
    y_fit = a*x + b
    ss_res = np.sum((y - y_fit)**2)
    ss_tot = np.sum((y - np.mean(y))**2)
    return 1 - ss_res/ss_tot

r2_1 = r_squared(x1, y1, a1, b1)
r2_2 = r_squared(x2, y2, a2, b2)

# 4) Representación gráfica
plt.figure(figsize=(8,5))
plt.plot(Temp_con, Amp_con, 'o-', label='Con espejo')
plt.plot(Temp_sin, Amp_sin, 's-', label='Sin espejo')

# Curvas de ajuste
x1_fit = np.linspace(x1.min(), x1.max(), 100)
x2_fit = np.linspace(x2.min(), x2.max(), 100)
plt.plot(x1_fit, a1*x1_fit + b1, '--', label=f'Ajuste lin. con espej. (R²={r2_1:.3f})')
plt.plot(x2_fit, a2*x2_fit + b2, '--', label=f'Ajuste lin. sin espej. (R²={r2_2:.3f})')

plt.xlabel('Temperatura (°C)')
plt.ylabel('Amplitud normalizada')
plt.title('Linealidad en la zona de emisión estimulada')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Impresión de parámetros de regresión
print(f'Con espejo: pendiente a1 = {a1:.4f}, intercepto b1 = {b1:.4f}, R² = {r2_1:.4f}')
print(f'Sin espejo: pendiente a2 = {a2:.4f}, intercepto b2 = {b2:.4f}, R² = {r2_2:.4f}')
```
