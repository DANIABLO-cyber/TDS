Este ejercicio plantea una transmisión óptica en fibra con pérdidas por atenuación de la fibra y por conectores. Se nos pide calcular la **potencia mínima de entrada** que garantice que a la salida haya **al menos 0,3 μW**. Vamos a resolverlo paso a paso **explicando todo en decibelios (dB)**:

---

### 🧾 Datos del enunciado:

* Longitud total de la fibra: **15 km**
* Atenuación de la fibra: **1,5 dB/km**
* Longitud por tramo de fibra: **1 km**, por lo tanto hay **14 conectores**
* Pérdida por conector: **0,8 dB**
* Potencia mínima de salida requerida: **P<sub>s</sub> = 0,3 μW**
* Se pide: **P<sub>e</sub> mínima**

---

### 🧮 Paso 1: Cálculo de pérdidas totales

#### 1. Pérdidas por conectores

Hay **14 conectores** → cada uno introduce **0,8 dB**:

$$
L_c = 14 \cdot 0{,}8 = 11{,}2\,\text{dB}
$$

#### 2. Pérdidas por la fibra

La atenuación es **1,5 dB/km** y hay **15 km**:

$$
L_f = 15 \cdot 1{,}5 = 22{,}5\,\text{dB}
$$

#### 3. Pérdida total del enlace:

$$
L_{\text{TOTAL}} = L_c + L_f = 11{,}2 + 22{,}5 = 33{,}7\,\text{dB}
$$

---

### 📐 Paso 2: Relación entre potencia de entrada y salida

Sabemos que:

$$
L_{\text{TOTAL}} = 10 \cdot \log_{10}\left(\frac{P_e}{P_s}\right)
$$

Despejamos:

$$
\frac{P_e}{P_s} = 10^{\frac{L_{\text{TOTAL}}}{10}} = 10^{\frac{33{,}7}{10}} \approx 23442{,}29
$$

Ahora multiplicamos:

$$
P_e = 23442{,}29 \cdot P_s = 23442{,}29 \cdot 0{,}3\,\mu\text{W} = 7032{,}7\,\text{nW} = 703{,}27\,\mu\text{W}
$$

---

### 💡 O también en unidades logarítmicas (dBm):

Sabemos que:

$$
P_{\text{dBm}} = 10 \log_{10}(P\, [\text{mW}])
$$

Entonces:

* Potencia de salida:

$$
P_s = 0{,}3\,\mu\text{W} = 0{,}0003\,\text{mW} \Rightarrow 10 \log(0{,}0003) \approx -35{,}2\,\text{dBm}
$$

* Potencia de entrada:

$$
P_e\, [\text{dBm}] = P_s\, [\text{dBm}] + L_{\text{TOTAL}} = -35{,}2 + 33{,}7 = -1{,}5\,\text{dBm}
$$

* Pasamos de nuevo a lineal:

$$
P_e = 10^{\frac{-1{,}5}{10}} = 0{,}707\,\text{mW} = 707\,\mu\text{W}
$$

(Es consistente con el valor anterior, con diferencia por redondeos.)

---

### ✅ Conclusión:

Para asegurar que a la salida haya **0,3 μW**, debemos inyectar una potencia de al menos:

$$
\boxed{703{,}27\,\mu\text{W} \quad \text{ó} \quad -1{,}5\,\text{dBm}}
$$

### **Datos del enunciado:**

* Potencia de salida: $P_s = 3\, \mu W$
* Longitud de la fibra: $L = 8\, km$

---

### **a) Atenuación total en dB**

La fórmula general de atenuación en dB es:

$$
\alpha_T = -10 \log\left( \frac{P_s}{P_e} \right)
$$

Sustituyendo:

$$
\alpha_T = -10 \log\left( \frac{3}{120} \right) = -10 \log(0.025) = 16\, \text{dB}
$$

> 📌 **Esto significa que se pierden 16 dB en total desde que entra hasta que sale la señal.**

---

### **b) Atenuación por kilómetro**

Ahora dividimos la atenuación total entre la longitud de la fibra:

$$
\alpha_T\ [\text{dB/km}] = \frac{16}{8} = 2\, \text{dB/km}
$$

> ✅ Esta es la **pérdida por unidad de longitud**, muy útil para estimar pérdidas en otros tramos.

---

### **c) Atenuación total con empalmes**

Nos dicen ahora que se trata de una fibra de 10 km de la misma clase, pero **con empalmes cada 1 km**. Cada empalme introduce una pérdida de **1 dB**, por tanto:

* Número de empalmes = 10 km → 9 empalmes
* Pérdida por empalmes: $9 \cdot 1 = 9\, \text{dB}$
* Atenuación por fibra: $2\, \text{dB/km} \cdot 10\, \text{km} = 20\, \text{dB}$

$$
\alpha_{\text{total}} = 9 + 20 = 29\, \text{dB}
$$

---

### **d) Razón de potencia de entrada/salida en el caso c)**

Queremos hallar la razón $\frac{P_e}{P_s}$ que produce una atenuación de 29 dB:

$$
\alpha = -10 \log\left( \frac{P_s}{P_e} \right) \Rightarrow \frac{P_s}{P_e} = 10^{-\frac{29}{10}} \Rightarrow \frac{P_e}{P_s} = 10^{2.9} \approx 794.3
$$

> 🔍 Esto significa que **la potencia de entrada es 794 veces mayor que la de salida**.

---

### **Resumen final:**

| b) Atenuación por km               | $2 \, \text{dB/km}$ |

Este ejercicio trata sobre el cálculo de pérdidas en una unión entre dos fibras ópticas con salto de índice, trabajando en tercera ventana (λ = 1550 nm). El enunciado especifica varios tipos de desalineación (lateral, longitudinal y angular), además de un pequeño “gap” de aire. Vamos a analizarlo paso por paso.

---

## 🧩 DATOS:

* **λ = 1550 nm**
* **n₁ = 1.47** (índice del núcleo)
* **Δ = 2 %** (diferencia relativa de índices)
* **d = 80 μm** → **a = 40 μm**
* **Desalineación lateral:** 2 μm
* **Desalineación longitudinal:** 4 μm
* **Desalineación angular:** 3°

---

## Paso 1: Pérdida por reflexión (Fresnel)

Dado que hay un pequeño gap de aire entre fibras, **sí hay pérdida Fresnel**, incluso si las fibras son iguales.

La pérdida se calcula con:

$$
L_F = -20 \log \left( \frac{4 n_1 n_0}{(n_1 + n_0)^2} \right)
$$

Con $n_1 = 1.47$ y $n_0 = 1$ (aire):

$$
L_F = -20 \log \left( \frac{4 \cdot 1.47 \cdot 1}{(1.47 + 1)^2} \right) \approx 0.3203 \ \text{dB}
$$

---

## Paso 2: Ver si es monomodo o multimodo

Para saber qué fórmulas usar, calculamos el parámetro $V$:

$$
V = \frac{2 \pi a}{\lambda} \cdot n_{nu} \cdot \sqrt{2 \Delta}
$$

$$
V = \frac{2 \pi \cdot 40 \cdot 10^{-6}}{1550 \cdot 10^{-9}} \cdot 1.47 \cdot \sqrt{2 \cdot 0.02} \approx 4.217
$$

Como $V > 2.405$, es **multimodo**.

---

## Paso 3: Pérdida por desalineación lateral

Usamos la fórmula exacta (válida para fibras multimodo):

$$
L_{lat} = -10 \log \left[ \frac{2}{\pi} \arccos \left( \frac{d}{2a} \right) - \frac{d}{6a} \left( 5 - \frac{d^2}{2a^2} \right) \sqrt{1 - \left( \frac{d}{2a} \right)^2} \right]
$$

Sustituyendo $d = 2 \ \mu m$, $a = 40 \ \mu m$:

$$
L_{lat} \approx 0.14047334 \ \text{dB}
$$

---

## Paso 4: Pérdida por desalineación longitudinal

Usamos la fórmula exacta para multimodo:

$$
L_{long} = -10 \log \left[ 1 - \frac{z}{a} \cdot \frac{2 n_e^2}{a \pi \text{AN}^2} \cdot \left( \arcsin\left( \frac{\text{AN}}{n_e} \right) - \frac{\text{AN}}{n_e} \sqrt{1 - \left( \frac{\text{AN}}{n_e} \right)^2} \right) \right]
$$

* $z = 4 \ \mu m$
* $a = 40 \ \mu m$
* $n_e = n_1 = 1.47$
* $\text{AN} = \sqrt{n_1^2 - n_2^2} \approx 0.294$

Resultado exacto:

$$
L_{long} = 0.056024125 \ \text{dB}
$$

Resultado aproximado:

$$
L_{long} \approx 0.054531093 \ \text{dB}
$$

🔹 Error relativo ≈ 2.7%

---

## Paso 5: Pérdida por desalineación angular

Fórmula exacta:

$$
L_{ang} = -10 \log \left[ \frac{2}{\pi} \arccos \left( \frac{n_e \sin(\theta/2)}{\text{AN}} \right) - \frac{n_e \sin(\theta/2)}{\text{AN}} \sqrt{1 - \left( \frac{n_e \sin(\theta/2)}{\text{AN}} \right)^2} \right]
$$

Con $\theta = 3^\circ$, $\text{AN} = 0.294$:

$$
L_{ang} = 0.5218203781 \ \text{dB}
$$

Aproximación:

$$
L_{ang} \approx 0.5225548833 \ \text{dB}
$$

🔹 Error relativo ≈ 0.15% (aceptable)

---

## Paso 6: Resultado total

### Usando fórmula exacta:

$$
L_{\text{total}} = L_F + L_{lat} + L_{long} + L_{ang} \approx 1.0386 \ \text{dB}
$$

### Aproximación:

$$
L_{\text{total}} \approx 1.0399 \ \text{dB}
$$

🔹 Error relativo total ≈ 0.07%

---

## ✅ Conclusión

* La pérdida total en la unión entre estas fibras, considerando todas las fuentes (Fresnel, desalineaciones), es de aproximadamente **1.039 dB**.
* Se puede usar la versión aproximada con muy poco error.
* Este tipo de análisis es clave para estimar la degradación de señal en enlaces largos y optimizar empalmes.
