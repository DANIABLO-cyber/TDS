### 1. ¿Para qué se coloca un atenuador de 15 dB delante del receptor en la práctica del EDFA?

* **Evitar la saturación del fotodetector**. El receptor lineal de la fuente-receptor del banco EDFA deja de trabajar correctamente a potencias ≳ 1 mW (0 dBm). El atenuador fija de forma segura la potencia de entrada \~15 dB por debajo, garantizando medida fiable del nivel de señal y de la ganancia del amplificador.&#x20;
* **Ampliar el rango dinámico de la práctica**. Con la atenuación se puede inyectar al EDFA potencias relativamente altas (para estudiar, p.ej., la saturación del propio amplificador) sin riesgo para el detector.
* **Simular pérdidas de enlace** de forma controlada, reproduciendo las atenuaciones que aparecerían en un enlace real de varios kilómetros de fibra.

### 2. Nd-YAG – primera parte: ¿de quién es la radiación que ve el detector?

En la fase inicial se retira el filtro IR y uno de los espejos para **medir exclusivamente la luz que emite el diodo láser de bombeo (≈ 808 nm)**; esta radiación es la que llega al fotodetector. Sólo cuando se coloca el filtro IR (λ > 1000 nm) se observa la emisión estimulada del Nd-YAG (1064 nm).

### 3. He-Ne – ¿importa mover el espejo móvil?

Sí importa, porque la separación L entre los dos espejos determina si la cavidad es estable y cuánta potencia se obtiene:

| Concepto                                                                                                                                                            | Consecuencia al variar **L**                                                                                                                                                                                           |
| ------------------------------------------------------------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Criterio de estabilidad** 0 ≤ g₁g₂ ≤ 1, con g₁ = 1-L/R₁, g₂ = 1-L/R₂ (R₁ → ∞ para el espejo plano). Para nuestro resonador plano-cóncavo se reduce a **L ≤ R₂**.  | Si L > R₂ el láser deja de oscilar; si  L ≪ R₂ la pérdida por difracción aumenta y la potencia baja.                                                                                                                   |
| **Espaciamiento modal** Δν = c/2nL                                                                                                                                  | Alargar L reduce la separación entre modos longitudinales y puede cambiar la estructura espectral.                                                                                                                     |
| **Umbral de ganancia**                                                                                                                                              | A mayor L, mayores pérdidas por difracción; el umbral de corriente crece y la potencia de salida disminuye. Los datos experimentales muestran el máximo alrededor de L ≈ 0,5 R₂ y una caída abrupta al acercarse a R₂. |

En el montaje del laboratorio se parte de un ajuste “óptimo” (\~45–70 cm). Mover el espejo fuera de ese intervalo sí altera la señal, aunque el espejo plano permanezca fijo.

### 4. Entrenador óptico – diferencias entre LED y láser más allá de la linealidad

| Aspecto                        | Láser de semiconductor                                                        | LED                                                                             |
| ------------------------------ | ----------------------------------------------------------------------------- | ------------------------------------------------------------------------------- |
| **Corriente umbral**           | Presenta Ith clara (15 mA en la práctica).                                    | No existe umbral; emisión continua y proporcional a la corriente.               |
| **Eficiencia y pendiente P-I** | Alta pendiente (≈0.11 W/A) ⇒ mayor eficacia de conversión.                    | Pendiente ≲ 0.01 W/A; mucha menos potencia acoplable.                           |
| **Anchura espectral**          | Muy estrecha (∼0.1 nm) → baja dispersión, admite datos ≥ 10 Gb/s.             | Ancho de ∼30–40 nm (gaussiano); limitado a < 200 Mb/s y enlaces cortos.         |
| **Divergencia del haz**        | Haz casi colimado ⇒ fácil acoplo a fibra monomodo.                            | Patrón casi isotrópico; se usa sobre todo con fibra multimodo.                  |
| **Coherencia / ruido**         | Luz coherente (fase y frecuencia) – permite sistemas interferométricos, DWDM. | Coherencia muy baja – inapropiado para interferómetros, pero inmune al speckle. |
| **Seguridad ocular**           | Mayor densidad de potencia; requiere precauciones de láser clase III/IV.      | Potencia baja y visible: riesgos mucho menores.                                 |
| **Sensibilidad térmica**       | λ aumenta ≈0.25 nm K⁻¹; exige control de temperatura en enlaces exigentes.    | Ancho espectral domina sobre los pequeños desplazamientos de λ.                 |

Estas diferencias explican por qué en los resultados del entrenador, además de la distinta linealidad, **el láser ofrece potencias mucho mayores, una respuesta más abrupta tras Ith y un haz fácilmente acoplable**, mientras que **el LED muestra salida más modesta, ancha en espectro y estable desde cero corriente**.

### 1 · ¿Para qué sirve el atenuador de 15 dB que colocamos delante del receptor en la práctica del EDFA?

Imagina que tu receptor óptico es como un micrófono muy sensible. Habla bajo y todo se entiende; grita demasiado cerca y el sonido se distorsiona.
El fotodiodo que tenemos en el banco EDFA “se distorsiona” (se satura) cuando le llega alrededor de 1 mW de potencia. Para que podamos **estudiar amplificaciones grandes sin dañar el detector ni perder la linealidad**, insertamos en la línea un pequeño pad de fibra que quita 15 dB (unas treinta veces menos intensidad). Así el EDFA puede entregar varios milivatios y, sin embargo, al “micrófono” solo le llega un susurro seguro y medible .

### 2 · Nd-YAG: en la primera parte solo vemos “una señal óptica” en el detector. ¿De dónde procede realmente?

Al empezar, la cavidad del Nd-YAG aún no está configurada como láser; simplemente iluminamos el cristal con **el propio diodo de bombeo** (≈ 808 nm).
Ese diodo actúa como linterna y parte de su luz atraviesa el cristal y los espejos hasta el fotodetector. Por eso lo que registras no es la emisión del Nd-YAG sino **la radiación de bombeo que no ha sido absorbida**. Más adelante, cuando añades el espejo faltante y -sobre todo- el filtro IR que deja pasar solo longitudes de onda > 1000 nm, empiezas a medir la luz láser de 1064 nm generada en el Nd-YAG propiamente dicho .

### 3 · He-Ne: ¿“Daría igual” mover el espejo cóncavo?

No, y aquí viene la comparación doméstica: piensa en dos espejos enfrentados que forman un pasillo de reflejos; si los separas demasiado o los juntas demasiado, el haz se “escapa” y se apaga la fiesta.
En nuestro resonador hay un espejo plano fijo y otro cóncavo móvil. El láser solo funciona mientras la distancia **L esté dentro de la zona de estabilidad** (básicamente L ≤ R, el radio de curvatura del espejo cóncavo). Si te acercas al borde de ese rango, la luz sigue saliendo pero cada vez con menos intensidad; si lo sobrepasas, desaparece el láser porque ya no se consiguen los múltiples rebotes necesarios para mantener la ganancia .
Por eso, aunque “el espejo derecho esté siempre fijo en su soporte”, desplazarlo un poco cambia la potencia y, si te excedes, apaga la oscilación.

### 4 · En el entrenador óptico, aparte de la linealidad, ¿qué otras diferencias práctica-notables hay entre usar un LED y un láser?

* **Arranque y umbral**
  El LED se comporta como una bombilla diminuta: cuanto más corriente, más luz, desde cero.
  El láser necesita superar una **corriente umbral** (≈ 15 mA en tu ensayo) para que aparezca luz coherente; antes de ese punto apenas brilla .

* **Cantidad y “calidad” de la luz**
  Una vez arrancado, el láser entrega mucha más potencia y con un espectro muy estrecho (pico casi monocromático). Esto permite mandar datos muy rápidos por una misma fibra sin que el pulso se “desparrame”.
  El LED emite un abanico de longitudes de onda bastante ancho y la intensidad total es menor; ideal para distancias cortas y aplicaciones menos exigentes .

* **Direccionalidad y acoplo a fibra**
  El haz del láser sale colimado, como un puntero; se acopla con facilidad a fibra monomodo. El LED envía luz en muchas direcciones y suele usarse con fibra multimodo, donde las pérdidas por mala alineación son menos críticas.

En resumen, el LED es sencillo y robusto, pero el láser -aunque más delicado- te permite obtener **más potencia, un punto de emisión muy definido y, tras su umbral, una respuesta más abrupta** que se ve claramente en las gráficas P-I de la práctica.
