Aquí tienes el desarrollo paso a paso para calcular el ensanchamiento del pulso en la fibra debido a la guía (dispersión de guía) y al material (dispersión cromática):

---

## 1. Datos del problema

* **Índice de refracción del núcleo**
  $n_1(\lambda) = 1.45084 \;-\;0.00334\,\lambda^2\;+\;\dfrac{0.00292}{\lambda^2}\quad$ con $\lambda$ en µm.

* **Diferencia relativa de índices**

  $$
    \Delta = \frac{n_1 - n_2}{n_1} = 0.02
    \quad\Longrightarrow\quad
    n_2 = n_1\,(1-\Delta)
  $$

* **Radio del núcleo**

  $$
    a = \frac{10\;\mu\text{m}}{2} = 5\;\mu\text{m}
  $$

* **Longitud de fibra**

  $$
    L = 3\;\text{km}
  $$

* **Ventana de operación**
  $\lambda_0 = 1.30\;\mu\text{m}$

* **Ancho espectral de la fuente**
  $\Delta\lambda = 100\;\text{nm} = 0.100\;\mu\text{m}$

* **Velocidad de la luz**

  $$
    c = 3\times10^8\;\mathrm{m/s}
         = 3\times10^5\;\mathrm{km/s}
         = 300\;\mu\mathrm{m/ps}
  $$

---

## 2. Dispersión de guía ($D_w$)

1. **Cálculo de $n_1$ y $n_2$ en $\lambda_0$:**

   $$
     n_1 = 1.45084 - 0.00334\,(1.30)^2 + \frac{0.00292}{(1.30)^2}
         \approx 1.4470
   $$

   $$
     n_2 = n_1\,(1 - 0.02) \approx 1.4470 \times 0.98 = 1.4181
   $$

2. **Número de guía $V$:**

   $$
     V = \frac{2\pi\,a\,n_1}{\lambda_0}\,\sqrt{2\Delta}
       = \frac{2\pi\,(5\,\mu\mathrm{m})\,(1.4470)}{1.30\,\mu\mathrm{m}}\,
         \sqrt{2\cdot0.02}
       \approx 7.0
   $$

3. **Parámetro**
   $\displaystyle V\,\frac{d^2}{dV^2}[\,b(V)\,]$
   Para una fibra convencional se usa la aproximación (véase diapo):

   $$
     V\frac{d^2}{dV^2}\bigl[b(V)\bigr]
     \;\approx\; 0.080 \;+\; 0.549\,\bigl(2.834 - V\bigr)^2
     \;\approx\; 0.080 + 0.549\,(2.834 - 7.0)^2
     \approx 9.6
   $$

4. **Coeficiente de dispersión de guía**

   $$
     D_w
     = -\,\frac{n_2\,\Delta}{c\,\lambda_0}\,
       \Bigl[V\,\frac{d^2b}{dV^2}\Bigr]
   $$

   Sustituyendo en unidades coherentes (con $c=300\;\mu\mathrm{m/ps}$, $\lambda_0$ en µm):

   $$
     D_w
     \approx -\,\frac{1.4181\;\times 0.02}{300\;\times 1.30}\;\times 9.6
     \approx -0.002\;\frac{\mathrm{ps}}{\mathrm{nm\cdot km}}
     \quad(\text{prácticamente }\pm 0.002)
   $$

   (**Nota**: el signo indica que la guía “compensa” ligeramente la dispersión del material.)

5. **Ensanchamiento por guía**

   $$
     \Delta\tau_w = \bigl|D_w\bigr|\;\times\;\Delta\lambda\;\times\;L
     \;=\; 0.002\;\tfrac{\mathrm{ps}}{\mathrm{nm\cdot km}}
           \;\times\;100\;\mathrm{nm}\;\times\;3\;\mathrm{km}
     \;\approx\;6\;\mathrm{ps}
   $$

---

## 3. Dispersión de material ($D_m$)

En la diapo se da la aproximación válida para vidrios comerciales en 1,25–1,66 µm:

$$
  D_m \;\approx\; 122\;\Bigl(1 - \frac{\lambda_{ZD}}{\lambda_0}\Bigr)
  \quad,\quad
  \lambda_{ZD} = 1.276\;\mu\text{m}  \text{longitud de onda de dispersión cero}
$$

$$
  D_m \approx 122\,\Bigl(1 - \frac{1.276}{1.300}\Bigr)
        = 122\,(1 - 0.9815)
        \approx 2.26\;\frac{\mathrm{ps}}{\mathrm{nm\cdot km}}
$$

**Ensanchamiento por material**

$$
  \Delta\tau_m = D_m\;\times\;\Delta\lambda\;\times\;L
  = 2.26\;\frac{\mathrm{ps}}{\mathrm{nm\cdot km}}
    \times 100\;\mathrm{nm}
    \times 3\;\mathrm{km}
  \;\approx\;678\;\mathrm{ps}
$$

---

## 4. Resultados y conclusiones

* **Ensanchamiento por guía**
  $\displaystyle \Delta\tau_w \approx 6\;\mathrm{ps}$

* **Ensanchamiento por material**
  $\displaystyle \Delta\tau_m \approx 678\;\mathrm{ps}$

* **Dispersión total**

  $$
    D_T = D_w + D_m \approx 2.26 - 0.002 \approx 2.26\;\tfrac{\mathrm{ps}}{\mathrm{nm\cdot km}}
    \quad\Longrightarrow\quad
    \Delta\tau_T = D_T\,\Delta\lambda\,L \approx 684\;\mathrm{ps}
  $$

Como ves, **la dispersión del material** domina por mucho (∼680 ps frente a ≲10 ps de la guía), por eso la ZMD (longitud de dispersión cero) se sitúa alrededor de 1,32 µm en una fibra monomodo convencional.
___

La **aproximación**:

$$
V \frac{d^2}{dV^2}(bV) \approx 0{,}080 + 0{,}549\, (2{,}834 - V)^2
$$

es una **expresión empírica** válida **solo para fibras monomodo** (single-mode fibers) en las que:

* $V \lesssim 2.405$ corresponde al **límite de corte** (por debajo de este valor la fibra es monomodo).
* Pero esta **aproximación es más amplia**, y se puede usar con buena precisión para valores de $V$ entre aproximadamente **1.5 y 4**, como se indica en varios textos de fibras ópticas (como Ghatak & Thyagarajan o Agrawal).

---

### ❗ En resumen, puedes usar esta fórmula cuando:

✅ La fibra es **monomodo**
✅ El número de guía $V$ está en el rango aproximado de **1.5 a 4** (idealmente ≤ 3)
✅ Estás evaluando **dispersión de guía** y necesitas calcular $D_w$

---

### Ejemplo típico:

En el ejercicio anterior, obtuvimos:

$$
V \approx 7
$$

Eso **está fuera del rango ideal** para esta aproximación. Por tanto:

* Si estás buscando una estimación rápida, **se puede usar con precaución**, entendiendo que introduce **error**.
* Para mayor precisión, deberías usar tablas numéricas de $b(V)$ o simulaciones, porque la fórmula ya no representa con fidelidad el comportamiento de la función para $V > 4$.

¿Quieres que repitamos ese paso con una mejor aproximación o te interesa ver gráficamente el error introducido por usarla fuera de rango?


Vamos a resolver punto por punto:

---

### **Datos comunes**

* Longitud de cavidad:

  $$
    L = 400\;\mu\mathrm m = 0{,}04\;\mathrm{cm}.
  $$
* Índice del semiconductor (AlGaAs):
  $n = 3{,}6$.
* Medio exterior: $n_{\rm ext}\approx1$.
* Coeficiente de absorción interna (pérdidas de material):
  $\displaystyle \alpha_{\rm int}=30\;\mathrm{cm^{-1}}$.
* Longitud de onda de emisión:
  $\displaystyle \lambda=850\;\mathrm{nm}$.

---

## a) Ganancia umbral sin recubrimientos

#### 1. Cálculo de la reflectividad de los extremos

Para extremidades “a hueso” (aire–semiconductor):

$$
R_1 = R_2
= \left(\frac{n - n_{\rm ext}}{n + n_{\rm ext}}\right)^2
= \left(\frac{3.6 - 1}{3.6 + 1}\right)^2
= 0.32
$$

#### 2. Pérdidas espejo (αₘ)

$$
\alpha_m
= \frac{1}{2L}\,\ln\frac{1}{R_1\,R_2}
= \frac{1}{2\cdot0.04}\;\ln\frac{1}{0.32^2}
$$

$$
\ln\frac{1}{0.32^2}=2\,\ln\frac1{0.32}=2\times1.139=2.278
\quad\Longrightarrow\quad
\alpha_m = \frac{2.278}{0.08}\approx28.5\;\mathrm{cm^{-1}}
$$

#### 3. Ganancia umbral total

$$
g_{\rm th}
=\alpha_{\rm int}+\alpha_m
=30 + 28.5
\approx58.5\;\mathrm{cm^{-1}}
$$

---

## b) Ganancia umbral con recubrimiento 80 % en cada extremo

Ahora $R_1=R_2=0.80$.

$$
\alpha_m
= \frac{1}{2\cdot0.04}\,\ln\frac{1}{0.8^2}
= \frac{\ln(1/0.64)}{0.08}
= \frac{0.4463}{0.08}
\approx5.58\;\mathrm{cm^{-1}}
$$

$$
g_{\rm th}
=30 + 5.58
\approx35.6\;\mathrm{cm^{-1}}
$$

---

## c) Eficiencia cuántica interna (ηᵢ), sabiendo que

$\displaystyle \frac{dP}{dI}=0.2\;\mathrm{W/A}$ **por un solo facet**

---

### 1. Relación entre pendiente experimental y eficiencia interna

El **pendiente de la curva P–I** en un facet viene dada por

$$
\frac{dP_1}{dI}
= \frac{h\nu}{e}\;\eta_d^{(1)}
$$

donde $\eta_d^{(1)}$ es la eficiencia cuántica diferencial que sale por ese facet.
Y a su vez

$$
\eta_d^{(1)}
= \eta_i \;\times\;
\frac{\alpha_m}{\,\alpha_{\rm int}+\alpha_m\,}
$$

de modo que

$$
\frac{dP_1}{dI}
= \frac{h\nu}{e}\;
\eta_i\;
\frac{\alpha_m}{\alpha_{\rm int}+\alpha_m}
\quad\Longrightarrow\quad
\eta_i
= \frac{\,\dfrac{dP_1}{dI}\,}{\,\dfrac{h\nu}{e}\,\dfrac{\alpha_m}{\alpha_{\rm int}+\alpha_m}\,}.
$$

---

### 2. Cálculo de $h\nu/e$

$$
h\nu = \frac{h\,c}{\lambda}
,\quad
\frac{h\nu}{e}
= \frac{1240\;\mathrm{eV\cdot nm}}{850\;\mathrm{nm}}
\approx1.459\;\mathrm{V}.
$$

---

### 3. Caso (a): $\alpha_m=28.5$, $\alpha_{\rm int}=30$

$$
\frac{\alpha_m}{\alpha_{\rm int}+\alpha_m}
= \frac{28.5}{30 + 28.5}
= 0.487
$$

$$
\eta_i^{(a)}
= \frac{0.2\;\mathrm{W/A}}
       {\,1.459\;\mathrm{V}\times0.487\,}
\approx0.28\;(28\%)
$$

---

### 4. Caso (b): $\alpha_m=5.58$, $\alpha_{\rm int}=30$

$$
\frac{\alpha_m}{\alpha_{\rm int}+\alpha_m}
= \frac{5.58}{30 + 5.58}
= 0.157
$$

$$
\eta_i^{(b)}
= \frac{0.2}
       {\,1.459\times0.157\,}
\approx0.87\;(87\%)
$$

---

## **Resumen de resultados**

| Caso                              | $g_{\rm th}$ \[cm⁻¹] | $\eta_i$ |
| --------------------------------- | -------------------- | -------- |
| **a)** sin recubrimiento          | 58.5                 | 28 %     |
| **b)** con R=80 % en cada extremo | 35.6                 | 87 %     |

---

**Interpretación:**
– Con refletores malos (caso a) pérdidas espejo ≈ internas ⇒ baja ηᵢ (\~28 %).
– Mejorar R reduce pérdidas espejo y por tanto sube ηᵢ hasta casi un 90 %.


A continuación tienes el procedimiento completo para calcular el ensanchamiento temporal ( τ ) que sufre un pulso al propagarse por una fibra, debido únicamente a la dispersión del material, y la sustitución paso a paso de las fórmulas, tanto para los casos de un LED con Δλ = 30 nm como para un láser con Δλ = 2 nm.

---

## 1. Fórmula básica de dispersión material

La **dispersión material** se caracteriza por el parámetro

$$
D_M \;=\;\frac{1}{L}\;\frac{\tau}{\Delta\lambda}
$$

con unidades típicas de ps/(nm·km).

De aquí despejamos el **ensanchamiento de pulso**:

$$
\tau \;=\; D_M \;\times\; L \;\times\; \Delta\lambda
$$

donde:

* $D_M$ = parámetro de dispersión material (ps/(nm·km)),
* $L$ = longitud de la fibra (km),
* $\Delta\lambda$ = anchura espectral de la fuente (nm),
* $\tau$ = ensanchamiento del pulso (ps).

---

## 2. Sustitución de valores

### Datos comunes

* Longitud de fibra:

  $$
   L = 80\;\text{km}
  $$
* Para λ = 850 nm:

  $$
   D_M = -105{,}5\;\frac{\text{ps}}{\text{nm km}}
  $$
* Para λ = 1300 nm:

  $$
   D_M = -2{,}8\;\frac{\text{ps}}{\text{nm km}}
  $$

---

### 2.1. LED con $\Delta\lambda = 30\;\text{nm}$

1. **Para λ = 850 nm**

   $$
   \tau_{850}
   = D_M \times L \times \Delta\lambda
   = -105{,}5\;\frac{\text{ps}}{\text{nm km}}
     \;\times\; 80\;\text{km}
     \;\times\; 30\;\text{nm}
   = -105{,}5 \times 2400\;\text{ps}
   = -253200\;\text{ps}
   = 253{,}2\;\text{ns}
   $$

2. **Para λ = 1300 nm**

   $$
   \tau_{1300}
   = -2{,}8\;\frac{\text{ps}}{\text{nm km}}
     \;\times\; 80\;\text{km}
     \;\times\; 30\;\text{nm}
   = -2{,}8 \times 2400\;\text{ps}
   = -6720\;\text{ps}
   = 6{,}72\;\text{ns}
   $$

> **Observación:** El signo «–» simplemente indica que la fibra en ese rango de longitudes de onda tiene dispersión que adelanta (en lugar de retrasar) el centro del pulso; a efectos de ancho de pulso tomamos el valor absoluto.

---

### 2.2. Láser con $\Delta\lambda = 2\;\text{nm}$

1. **Para λ = 850 nm**

   $$
   \tau_{850}
   = -105{,}5\;\frac{\text{ps}}{\text{nm km}}
     \;\times\; 80\;\text{km}
     \;\times\; 2\;\text{nm}
   = -105{,}5 \times 160\;\text{ps}
   = -16880\;\text{ps}
   = 16{,}88\;\text{ns}
   $$

2. **Para λ = 1300 nm**

   $$
   \tau_{1300}
   = -2{,}8\;\frac{\text{ps}}{\text{nm km}}
     \;\times\; 80\;\text{km}
     \;\times\; 2\;\text{nm}
   = -2{,}8 \times 160\;\text{ps}
   = -448\;\text{ps}
   = 0{,}448\;\text{ns}
   $$

---

## 3. Interpretación y “segunda ventana”

* Para **anchura espectral grande** (LED, 30 nm) el ensanchamiento es de **200–250 ns** en λ=850 nm y de **6–7 ns** en λ=1300 nm.
* Para **anchura espectral extremadamente estrecha** (láser, 2 nm) baja a **<20 ns** en λ=850 nm y **<0,5 ns** en λ=1300 nm.

Es justamente en la **segunda ventana** (λ≈1300 nm) donde el parámetro de dispersión material es muy pequeño (solo −2,8 ps/(nm·km)), lo que reduce drásticamente el ensanchamiento temporal del pulso y **permite transmisiones a mayor velocidad y distancia sin limitarse tanto por la dispersión cromática**.

---

**Resumen numérico**

|    Fuente |    λ    | Δλ (nm) | DM (ps/(nm·km)) |  τ (ns)  |
| --------: | :-----: | :-----: | :-------------: | :------: |
|   **LED** |  850 nm |    30   |      −105,5     | 253,2 ns |
|   **LED** | 1300 nm |    30   |       −2,8      |  6,72 ns |
| **Láser** |  850 nm |    2    |      −105,5     | 16,88 ns |
| **Láser** | 1300 nm |    2    |       −2,8      | 0,448 ns |

---

De esta forma tienes tanto la deducción de la fórmula como la sustitución detallada paso a paso, apoyada en los datos de dispersión material especificados para cada longitud de onda.
