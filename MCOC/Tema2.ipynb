Aqu tienes el desarrollo paso a paso para calcular el ensanchamiento del pulso en la fibra debido a la gua (dispersin de gua) y al material (dispersin cromtica):

---

## 1. Datos del problema

* **ndice de refraccin del ncleo**
  $n_1(\lambda) = 1.45084 \;-\;0.00334\,\lambda^2\;+\;\dfrac{0.00292}{\lambda^2}\quad$ con $\lambda$ en m.

* **Diferencia relativa de ndices**

  $$
    \Delta = \frac{n_1 - n_2}{n_1} = 0.02
    \quad\Longrightarrow\quad
    n_2 = n_1\,(1-\Delta)
  $$

* **Radio del ncleo**

  $$
    a = \frac{10\;\mu\text{m}}{2} = 5\;\mu\text{m}
  $$

* **Longitud de fibra**

  $$
    L = 3\;\text{km}
  $$

* **Ventana de operacin**
  $\lambda_0 = 1.30\;\mu\text{m}$

* **Ancho espectral de la fuente**
  $\Delta\lambda = 100\;\text{nm} = 0.100\;\mu\text{m}$

* **Velocidad de la luz**

  $$
    c = 3\times10^8\;\mathrm{m/s}
         = 3\times10^5\;\mathrm{km/s}
         = 300\;\mu\mathrm{m/ps}
  $$

---

## 2. Dispersin de gua ($D_w$)

1. **Clculo de $n_1$ y $n_2$ en $\lambda_0$:**

   $$
     n_1 = 1.45084 - 0.00334\,(1.30)^2 + \frac{0.00292}{(1.30)^2}
         \approx 1.4470
   $$

   $$
     n_2 = n_1\,(1 - 0.02) \approx 1.4470 \times 0.98 = 1.4181
   $$

2. **Nmero de gua $V$:**

   $$
     V = \frac{2\pi\,a\,n_1}{\lambda_0}\,\sqrt{2\Delta}
       = \frac{2\pi\,(5\,\mu\mathrm{m})\,(1.4470)}{1.30\,\mu\mathrm{m}}\,
         \sqrt{2\cdot0.02}
       \approx 7.0
   $$

3. **Parmetro**
   $\displaystyle V\,\frac{d^2}{dV^2}[\,b(V)\,]$
   Para una fibra convencional se usa la aproximacin (vase diapo):

   $$
     V\frac{d^2}{dV^2}\bigl[b(V)\bigr]
     \;\approx\; 0.080 \;+\; 0.549\,\bigl(2.834 - V\bigr)^2
     \;\approx\; 0.080 + 0.549\,(2.834 - 7.0)^2
     \approx 9.6
   $$

4. **Coeficiente de dispersin de gua**

   $$
     D_w
     = -\,\frac{n_2\,\Delta}{c\,\lambda_0}\,
       \Bigl[V\,\frac{d^2b}{dV^2}\Bigr]
   $$

   Sustituyendo en unidades coherentes (con $c=300\;\mu\mathrm{m/ps}$, $\lambda_0$ en m):

   $$
     D_w
     \approx -\,\frac{1.4181\;\times 0.02}{300\;\times 1.30}\;\times 9.6
     \approx -0.002\;\frac{\mathrm{ps}}{\mathrm{nm\cdot km}}
     \quad(\text{prcticamente }\pm 0.002)
   $$

   (**Nota**: el signo indica que la gua compensa ligeramente la dispersin del material.)

5. **Ensanchamiento por gua**

   $$
     \Delta\tau_w = \bigl|D_w\bigr|\;\times\;\Delta\lambda\;\times\;L
     \;=\; 0.002\;\tfrac{\mathrm{ps}}{\mathrm{nm\cdot km}}
           \;\times\;100\;\mathrm{nm}\;\times\;3\;\mathrm{km}
     \;\approx\;6\;\mathrm{ps}
   $$

---

## 3. Dispersin de material ($D_m$)

En la diapo se da la aproximacin vlida para vidrios comerciales en 1,251,66 m:

$$
  D_m \;\approx\; 122\;\Bigl(1 - \frac{\lambda_{ZD}}{\lambda_0}\Bigr)
  \quad,\quad
  \lambda_{ZD} = 1.276\;\mu\text{m}  \text{longitud de onda de dispersin cero}
$$

$$
  D_m \approx 122\,\Bigl(1 - \frac{1.276}{1.300}\Bigr)
        = 122\,(1 - 0.9815)
        \approx 2.26\;\frac{\mathrm{ps}}{\mathrm{nm\cdot km}}
$$

**Ensanchamiento por material**

$$
  \Delta\tau_m = D_m\;\times\;\Delta\lambda\;\times\;L
  = 2.26\;\frac{\mathrm{ps}}{\mathrm{nm\cdot km}}
    \times 100\;\mathrm{nm}
    \times 3\;\mathrm{km}
  \;\approx\;678\;\mathrm{ps}
$$

---

## 4. Resultados y conclusiones

* **Ensanchamiento por gua**
  $\displaystyle \Delta\tau_w \approx 6\;\mathrm{ps}$

* **Ensanchamiento por material**
  $\displaystyle \Delta\tau_m \approx 678\;\mathrm{ps}$

* **Dispersin total**

  $$
    D_T = D_w + D_m \approx 2.26 - 0.002 \approx 2.26\;\tfrac{\mathrm{ps}}{\mathrm{nm\cdot km}}
    \quad\Longrightarrow\quad
    \Delta\tau_T = D_T\,\Delta\lambda\,L \approx 684\;\mathrm{ps}
  $$

Como ves, **la dispersin del material** domina por mucho (680 ps frente a 10 ps de la gua), por eso la ZMD (longitud de dispersin cero) se sita alrededor de 1,32 m en una fibra monomodo convencional.

$$

es una **expresin emprica** vlida **solo para fibras monomodo** (single-mode fibers) en las que:

* $V \lesssim 2.405$ corresponde al **lmite de corte** (por debajo de este valor la fibra es monomodo).
* Pero esta **aproximacin es ms amplia**, y se puede usar con buena precisin para valores de $V$ entre aproximadamente **1.5 y 4**, como se indica en varios textos de fibras pticas (como Ghatak & Thyagarajan o Agrawal).

## a) Ganancia umbral sin recubrimientos

#### 1. Clculo de la reflectividad de los extremos

Para extremidades a hueso (airesemiconductor):

$$
R_1 = R_2

= \left(\frac{3.6 - 1}{3.6 + 1}\right)^2
= 0.32
$$

#### 2. Prdidas espejo ()

$$
\alpha_m
= \frac{1}{2L}\,\ln\frac{1}{R_1\,R_2}

$$

## c) Eficiencia cuntica interna (), sabiendo que

$\displaystyle \frac{dP}{dI}=0.2\;\mathrm{W/A}$ **por un solo facet**

### 1. Relacin entre pendiente experimental y eficiencia interna

El **pendiente de la curva PI** en un facet viene dada por

$$
\frac{dP_1}{dI}
= \frac{h\nu}{e}\;\eta_d^{(1)}
$$

donde $\eta_d^{(1)}$ es la eficiencia cuntica diferencial que sale por ese facet.
Y a su vez

$$
\eta_d^{(1)}
= \eta_i \;\times\;
\frac{\alpha_m}{\,\alpha_{\rm int}+\alpha_m\,}
$$

de modo que

$$
\frac{dP_1}{dI}
= \frac{h\nu}{e}\;
\eta_i\;
\frac{\alpha_m}{\alpha_{\rm int}+\alpha_m}
\quad\Longrightarrow\quad
\eta_i
= \frac{\,\dfrac{dP_1}{dI}\,}{\,\dfrac{h\nu}{e}\,\dfrac{\alpha_m}{\alpha_{\rm int}+\alpha_m}\,}.

A continuacin tienes el procedimiento completo para calcular el ensanchamiento temporal (  ) que sufre un pulso al propagarse por una fibra, debido nicamente a la dispersin del material, y la sustitucin paso a paso de las frmulas, tanto para los casos de un LED con  = 30 nm como para un lser con  = 2 nm.

---

## 1. Frmula bsica de dispersin material

La **dispersin material** se caracteriza por el parmetro

$$
D_M \;=\;\frac{1}{L}\;\frac{\tau}{\Delta\lambda}
$$

con unidades tpicas de ps/(nmkm).

De aqu despejamos el **ensanchamiento de pulso**:

$$
\tau \;=\; D_M \;\times\; L \;\times\; \Delta\lambda
$$

donde:

* $D_M$ = parmetro de dispersin material (ps/(nmkm)),
* $L$ = longitud de la fibra (km),
* $\Delta\lambda$ = anchura espectral de la fuente (nm),
* $\tau$ = ensanchamiento del pulso (ps).

---

## 2. Sustitucin de valores

### Datos comunes

* Longitud de fibra:

  $$
   L = 80\;\text{km}
  $$
* Para  = 850 nm:

  $$
   D_M = -105{,}5\;\frac{\text{ps}}{\text{nm km}}
  $$
* Para  = 1300 nm:

  $$
   D_M = -2{,}8\;\frac{\text{ps}}{\text{nm km}}
  $$

---

### 2.1. LED con $\Delta\lambda = 30\;\text{nm}$

1. **Para  = 850 nm**

   $$
   \tau_{850}
   = D_M \times L \times \Delta\lambda
   = -105{,}5\;\frac{\text{ps}}{\text{nm km}}
     \;\times\; 80\;\text{km}
     \;\times\; 30\;\text{nm}
   = -105{,}5 \times 2400\;\text{ps}
   = -253200\;\text{ps}
   = 253{,}2\;\text{ns}
   $$

2. **Para  = 1300 nm**

   $$
   \tau_{1300}
   = -2{,}8\;\frac{\text{ps}}{\text{nm km}}
     \;\times\; 80\;\text{km}
     \;\times\; 30\;\text{nm}
   = -2{,}8 \times 2400\;\text{ps}
   = -6720\;\text{ps}
   = 6{,}72\;\text{ns}
   $$

> **Observacin:** El signo  simplemente indica que la fibra en ese rango de longitudes de onda tiene dispersin que adelanta (en lugar de retrasar) el centro del pulso; a efectos de ancho de pulso tomamos el valor absoluto.

---

### 2.2. Lser con $\Delta\lambda = 2\;\text{nm}$

1. **Para  = 850 nm**

   $$
   \tau_{850}
   = -105{,}5\;\frac{\text{ps}}{\text{nm km}}
     \;\times\; 80\;\text{km}
     \;\times\; 2\;\text{nm}
   = -105{,}5 \times 160\;\text{ps}
   = -16880\;\text{ps}
   = 16{,}88\;\text{ns}
   $$

2. **Para  = 1300 nm**

   $$
   \tau_{1300}
   = -2{,}8\;\frac{\text{ps}}{\text{nm km}}
     \;\times\; 80\;\text{km}
     \;\times\; 2\;\text{nm}
   = -2{,}8 \times 160\;\text{ps}
   = -448\;\text{ps}
   = 0{,}448\;\text{ns}
   $$

---

## 3. Interpretacin y segunda ventana

* Para **anchura espectral grande** (LED, 30 nm) el ensanchamiento es de **200250 ns** en =850 nm y de **67 ns** en =1300 nm.
* Para **anchura espectral extremadamente estrecha** (lser, 2 nm) baja a **<20 ns** en =850 nm y **<0,5 ns** en =1300 nm.

Es justamente en la **segunda ventana** (1300 nm) donde el parmetro de dispersin material es muy pequeo (solo 2,8 ps/(nmkm)), lo que reduce drsticamente el ensanchamiento temporal del pulso y **permite transmisiones a mayor velocidad y distancia sin limitarse tanto por la dispersin cromtica**.
