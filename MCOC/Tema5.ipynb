## a) Ganancia umbral $g_{\text{th}}$

En un lser de cavidad lineal

$$
\Gamma\,g_{\text{th}} = \alpha_{\text{int}} + \alpha_{\text{m}},
$$

donde el trmino de espejos (prdidas externas por unidad de longitud) es

$$
\boxed{\displaystyle
\alpha_{\text{m}}=\frac{1}{2L}\,\ln\!\left(\frac{1}{R_{1}R_{2}}\right)} .
$$

Como slo conocemos la **relacin** de reflectividades, tomamos el caso ms habitual de diodo lser: la cara trasera est prcticamente metalizada ($R_{2}\approx1$).
Entonces $R_{1}=0.525$ y

$$
\alpha_{\text{m}}=\frac{1}{2(3.0\times10^{-4})}\,
                 \ln\!\left(\frac{1}{0.525}\right)
              = 1.07\times10^{3}\,\text{m}^{-1}
              = 10.7\;\text{cm}^{-1}.
$$

Sustituyendo:

$$
g_{\text{th}}=\frac{\alpha_{\text{int}}+\alpha_{\text{m}}}{\Gamma}
            =\frac{30+10.7}{0.8}
            \approx\boxed{5.1\times10^{1}\;\text{cm}^{-1}}.
$$

*(Si se usa otro recubrimiento para la cara trasera, el valor cambia; el procedimiento es idntico).*

## b) Es buen lser?  Eficiencia cuntica total

La **eficiencia cuntica diferencial externa** (fraccin de electrones convertidos en fotones de salida) se escribe

$$
\eta_{\text{d}} = \eta_{i}\,
                  \frac{\alpha_{\text{m}}}{\alpha_{\text{m}}+\alpha_{\text{int}}},
$$

siendo $\eta_{i}$ la eficiencia cuntica **interna** (probabilidad de recombinacin radiativa).  Con los nmeros anteriores

$$
\frac{\alpha_{\text{m}}}{\alpha_{\text{m}}+\alpha_{\text{int}}}
        =\frac{10.7}{10.7+30}=0.263.
$$

Aun suponiendo el caso ideal $\eta_{i}=1$,

$$
\eta_{\text{d}}\le 26\%.
$$

En diodos modernos de telecom (InGaAsP o AlGaInAs) se esperan valores externos $\gtrsim 60\%$ y umbrales de ganancia del orden de $20{-}30\;\text{cm}^{-1}$. Con una ganancia umbral de 51 cm$^{-1}$ y $\eta_{\text{d}}\!\approx\!26\%$, **este dispositivo no sera considerado especialmente eficiente**:

* El umbral de ganancia es relativamente alto  ms corriente para encender.
* Slo \~ de los portadores inyectados se convierten en fotones tiles, incluso en el mejor escenario ( $\eta_{i}=1$ ).

> **Conclusin:** desde el punto de vista de eficiencia cuntica total, el lser es aceptable para fines docentes pero no competitivo con diodos comerciales de altas prestaciones.
