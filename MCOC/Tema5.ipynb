## a) Ganancia umbral $g_{\text{th}}$

En un láser de cavidad lineal

$$
\Gamma\,g_{\text{th}} = \alpha_{\text{int}} + \alpha_{\text{m}},
$$

donde el término de espejos (pérdidas externas por unidad de longitud) es

$$
\boxed{\displaystyle
\alpha_{\text{m}}=\frac{1}{2L}\,\ln\!\left(\frac{1}{R_{1}R_{2}}\right)} .
$$

Como sólo conocemos la **relación** de reflectividades, tomamos el caso más habitual de diodo láser: la cara “trasera” está prácticamente metalizada ($R_{2}\approx1$).
Entonces $R_{1}=0.525$ y

$$
\alpha_{\text{m}}=\frac{1}{2(3.0\times10^{-4})}\,
                 \ln\!\left(\frac{1}{0.525}\right)
              = 1.07\times10^{3}\,\text{m}^{-1}
              = 10.7\;\text{cm}^{-1}.
$$

Sustituyendo:

$$
g_{\text{th}}=\frac{\alpha_{\text{int}}+\alpha_{\text{m}}}{\Gamma}
            =\frac{30+10.7}{0.8}
            \approx\boxed{5.1\times10^{1}\;\text{cm}^{-1}}.
$$

*(Si se usa otro recubrimiento para la cara trasera, el valor cambia; el procedimiento es idéntico).*

## b) ¿Es “buen” láser? – Eficiencia cuántica total

La **eficiencia cuántica diferencial externa** (fracción de electrones convertidos en fotones de salida) se escribe

$$
\eta_{\text{d}} = \eta_{i}\,
                  \frac{\alpha_{\text{m}}}{\alpha_{\text{m}}+\alpha_{\text{int}}},
$$

siendo $\eta_{i}$ la eficiencia cuántica **interna** (probabilidad de recombinación radiativa).  Con los números anteriores

$$
\frac{\alpha_{\text{m}}}{\alpha_{\text{m}}+\alpha_{\text{int}}}
        =\frac{10.7}{10.7+30}=0.263.
$$

Aun suponiendo el caso ideal $\eta_{i}=1$,

$$
\eta_{\text{d}}\le 26\%.
$$

En diodos modernos de telecom (InGaAsP o AlGaInAs) se esperan valores externos $\gtrsim 60\%$ y umbrales de ganancia del orden de $20{-}30\;\text{cm}^{-1}$. Con una ganancia umbral de 51 cm$^{-1}$ y $\eta_{\text{d}}\!\approx\!26\%$, **este dispositivo no sería considerado especialmente eficiente**:

* El umbral de ganancia es relativamente alto → más corriente para encender.
* Sólo \~¼ de los portadores inyectados se convierten en fotones útiles, incluso en el mejor escenario ( $\eta_{i}=1$ ).

> **Conclusión:** desde el punto de vista de eficiencia cuántica total, el láser es aceptable para fines docentes pero no competitivo con diodos comerciales de altas prestaciones.

### Qué representa cada magnitud

| Símbolo               | Significado físico                                                                 |
| --------------------- | ---------------------------------------------------------------------------------- |
| $g_{\text{th}}$       | Ganancia modal necesaria para que la amplificación compense todas las pérdidas.    |
| $\Gamma$              | Factor de confinamiento: qué fracción del modo óptico transita por la zona activa. |
| $\alpha_{\text{int}}$ | Pérdidas internas (absorción y dispersión en la guía).                             |
| $\alpha_{\text{m}}$   | Pérdidas externas por transmisión en los espejos de la cavidad.                    |
| $\tau_{\text{ph}}$    | Tiempo medio que un fotón permanece en la cavidad.                                 |
| $S$                   | Número de fotones intracavidad en régimen estacionario.                            |
| $\eta_{i}$            | Eficiencia cuántica interna (probabilidad de emisión espontánea → fotón útil).     |
| $\eta_{\text{d}}$     | Eficiencia cuántica diferencial externa (pendiente $dP/dI$).                       |

Si se dispusiera de $\eta_{i}$ medida y de la corriente umbral exacta, podría refinarse la evaluación; con los datos facilitados bastan los cálculos anteriores para calificar el láser.
