```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
fig, ax = plt.subplots()
f = 1.0     # Frecuencia de la señal en Hz
fs = 5.0    # Frecuencia de muestreo en Hz (al menos 2*f por teorema de muestreo)

# Rango de tiempo: desde -1 a 1 con pasos de 1/fs
t = np.arange(-1, 1 + 1/fs, 1/fs)

# Señal: senoide
x = np.sin(2 * np.pi * f * t)

# Gráfico con círculos conectados por líneas
plt.plot(t, x, 'o-')

# Etiquetas con fuente grande
plt.xlabel('Time', fontsize=18)
plt.ylabel('Amplitude', fontsize=18)

plt.grid(True)
plt.show()
```
