```python
# Experiment 1: Generation of sinusoidal signal

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Defining the independent variable (tiempo)
t = np.linspace(0, 1, 1000)

# Step 2: Parámetros de la señal sinusoidal
A = 5     # Amplitud (V)
f = 5     # Frecuencia (Hz)
ph = 0    # Fase (radianes)

# Step 3: Ecuación de la señal
x = A * np.sin(2 * np.pi * f * t + ph)

# Step 4: Graficar la señal
plt.plot(t, x)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.title(f'A={A} V, F={f} Hz, $\\phi={ph}^{{\\circ}}$')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Generación de onda sinusoidal con diferentes fases

import numpy as np
import matplotlib.pyplot as plt

# Tiempo
t = np.linspace(0, 1, 100)

# Parámetros de la señal
A = 1              # Amplitud
f = 5              # Frecuencia en Hz
phi = [0, 90, 180, 270]  # Fases en grados

# Graficar
plt.figure(figsize=(10, 6))

for i in range(len(phi)):
    # Convertir fase a radianes y generar señal
    x = A * np.sin(2 * np.pi * f * t + phi[i] * np.pi / 180)

    # Crear subgráfico
    plt.subplot(2, 2, i + 1)
    plt.plot(t, x)
    plt.xlabel("Time (t)")
    plt.ylabel("Amplitude (V)")
    plt.title(r'$\Phi = {}^\circ$'.format(phi[i]))
    plt.grid(True)

plt.tight_layout()
plt.show()
```
