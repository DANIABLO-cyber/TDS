```python
import matplotlib.pyplot as plt
import numpy as np

#! TAREA 1
# Datos extra√≠dos de la tabla
distancia_cm = np.array([20, 30, 40, 50, 60, 70, 80, 90])
voltaje_mV = np.array([435, 265, 195, 140, 120, 93, 72, 65])

# C√°lculo de 1/sqrt(V)
inv_sqrt_V = 1 / np.sqrt(voltaje_mV)

# Ajuste lineal para comparaci√≥n
coef = np.polyfit(distancia_cm, inv_sqrt_V, 1)
fit_line = np.poly1d(coef)

# Gr√°fica
plt.figure(figsize=(8, 5))
plt.plot(distancia_cm, inv_sqrt_V, 'o', label='Datos experimentales', color='blue')
plt.plot(distancia_cm, fit_line(distancia_cm), '-', label='Ajuste lineal', color='red')

plt.xlabel("Distancia (cm)")
plt.ylabel(r"$\frac{1}{\sqrt{V}}$ (1/$\sqrt{\mathrm{Voltaje}}$)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# √Ångulos en grados
angles_deg = np.array([
    -60, -50, -40, -30, -20, -10, 0, 10, 20, 30, 40, 50, 60
])
angles_rad = np.radians(angles_deg)

# Datos con obst√°culo (en dB)
e_plane_db_obs = np.array([
    -62.36, -56.81, -59.16, -37.77, -29.86, -21.68, -18.60,
    -21.96, -29.21, -39.18, -58.00, -58.16, -63.63
])

h_plane_db_obs = np.array([
    -54.60, -46.05, -42.71, -37.64, -26.76, -19.33, -17.82,
    -20.99, -31.54, -34.43, -40.22, -52.17, -55.30
])

# Datos sin obst√°culo (en dB)
e_plane_db_clear = np.array([
    -44.35, -36.99, -31.13, -27.09, -15.70, -5.35, 0,
    -6.47, -17.97, -28.20, -31.38, -34.34, -43.65
])

h_plane_db_clear = np.array([
    -32.18, -25.38, -23.86, -21.45, -14.70, -6.34, -2.45,
    -6.07, -17.34, -22.45, -23.93, -28.18, -32.93
])

# Normalizar cada conjunto a su propio m√°ximo
e_plane_db_obs_norm = e_plane_db_obs - np.max(e_plane_db_obs)
h_plane_db_obs_norm = h_plane_db_obs - np.max(h_plane_db_obs)
e_plane_db_clear_norm = e_plane_db_clear - np.max(e_plane_db_clear)
h_plane_db_clear_norm = h_plane_db_clear - np.max(h_plane_db_clear)

# Crear figura con dos subplots en columnas (polar)
fig, (ax1, ax2) = plt.subplots(1, 2, subplot_kw={'projection': 'polar'}, figsize=(12, 6))

# Gr√°fico con obst√°culo
ax1.plot(angles_rad, e_plane_db_obs_norm, label='Plano E', color='blue', marker='o')
ax1.plot(angles_rad, h_plane_db_obs_norm, label='Plano H', color='red', marker='x')
ax1.set_theta_zero_location('N')
ax1.set_theta_direction(-1)
ax1.set_rlabel_position(135)
ax1.set_rticks([-60, -50, -40, -30, -20, -10, 0])
ax1.grid(True)
ax1.legend(loc='upper right')

# Gr√°fico sin obst√°culo
ax2.plot(angles_rad, e_plane_db_clear_norm, label='Plano E', color='blue', marker='o')
ax2.plot(angles_rad, h_plane_db_clear_norm, label='Plano H', color='red', marker='x')
ax2.set_theta_zero_location('N')
ax2.set_theta_direction(-1)
ax2.set_rlabel_position(135)
ax2.set_rticks([-60, -50, -40, -30, -20, -10, 0])
ax2.grid(True)
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
